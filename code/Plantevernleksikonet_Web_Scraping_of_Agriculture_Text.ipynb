# Download Webpages for different topics

Webpages for each topic can be reached by URL of the following form:


```
https://www.plantevernleksikonet.no/l/oppslag/number/
```
where "number" is a number in a certain range. For example, the following URL is a URL for the Småkattost webpage

```
https://www.plantevernleksikonet.no/l/oppslag/1700/
```

We can go through all numbers in a range from 0 to 10000, download webpages and check if they are not epmty




````python
url_pattern = "https://www.plantevernleksikonet.no/l/oppslag/"

topics_urls_start = 0
topics_urls_end = 3000

stopping_threshold_by_cons_empty_pages = 100

webpage_not_found_message = '404 - fant ikke siden'
````

````python
active_webpages = {}
active_webpage_numbers = []
n_cons_empty_pages = 0

start_time = time.time()
for number in range(topics_urls_start, topics_urls_end):

    topic_url = url_pattern + str(number) + '/'

    topic_page = requests.get(topic_url)
    topic_page.encoding = topic_page.apparent_encoding
    topic_webpage = BeautifulSoup(topic_page.text, 'html')

    empty_page_string_list = topic_webpage.head.find(string = webpage_not_found_message)

    if empty_page_string_list:
        n_cons_empty_pages += 1
    else:
        n_cons_empty_pages = 0
        active_webpages[number] = topic_webpage
        active_webpage_numbers.append(number)

    if n_cons_empty_pages == stopping_threshold_by_cons_empty_pages:
        print(f'breaking at number {number}')
        break

print(f'execution time: {(time.time() - start_time) / 60} minutes')
````

We can parse it by searching for specific html hashtags


Exception: some web pages had headers but no text content. Examples are:

*   Melon: https://www.plantevernleksikonet.no/l/oppslag/679/
*   Secoviridae: https://www.plantevernleksikonet.no/l/oppslag/1783/
*   Lyngfamilien: https://www.plantevernleksikonet.no/l/oppslag/408/

In that case they show the following message:

```
'Denne organismen mangler oppslag i leksikonet. Beklager!'
```

For example the webpage below:



````python
active_webpages[1783]
````

````python
not_in_encyclopedia_message = 'Denne organismen mangler oppslag i leksikonet. Beklager!'

webpages_content = {}

for number in active_webpage_numbers:
    webpage = active_webpages[number]

    not_in_encyclopedia = webpage.find(
        string = not_in_encyclopedia_message
        )

    if not_in_encyclopedia:
        continue

    title = webpage.head.find('title').getText().strip()

    headers_div = webpage.find(attrs = {'id' : 'entryHeader'})
    headers = []
    if headers_div:
        for h in headers_div.find_all(re.compile('^h[1-6]$')):
            headers.append(h.getText().strip())
        headers = '\n '.join(headers)

    summary = webpage.find(attrs = {'id' : 'summary'})
    if summary:
        summary = summary.getText().strip()

    webpage_dict = {}
    webpage_dict['page_number'] = number
    webpage_dict['headers'] = headers
    webpage_dict['summary'] = summary

    webpage_content = []
    content_headers = webpage.find_all('h2', { 'id': re.compile(r'chapter_')})
    for content_header in content_headers:
        content_header_text = content_header.getText().strip()

        if content_header_text == 'Litteratur':
            continue

        content = []
        for paragraph in content_header.find_next_siblings():

            if paragraph.name != 'p':
                break

            paragraph_text = paragraph.get_text() if paragraph is not None else ''
            content.append(paragraph_text)

        content = '\n '.join(content)
        webpage_content.append(content_header_text)
        webpage_content.append(content)

    webpage_content = '\n '.join(webpage_content)
    webpage_dict['content'] = webpage_content

    # if summary:
    webpages_content[title] = webpage_dict
````
